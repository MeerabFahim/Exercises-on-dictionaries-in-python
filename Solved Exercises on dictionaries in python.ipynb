```python
# Create a dictionary to store information about a person
person = {
    "name": "John Doe",
    "age": 30,
    "city": "New York",
    "hobbies": ["reading", "traveling", "photography"]
}
print("Complete Information:")
print(person)

# Access specific information
print("\nAccessing Specific Information:")
print(f"Name: {person['name']}")
print(f"Age: {person['age']}")
print(f"City: {person['city']}")
print(f"Hobbies: {', '.join(person['hobbies'])}")

# Modify the person's age
person["age"] = 31
print("\nUpdated Age:")
print(person["age"])

# Add a new key-value pair for the profession
person["profession"] = "Software Engineer"
print("\nAdded Profession:")
print(person["profession"])

# Display the updated dictionary
print("\nUpdated Complete Information:")
print(person)
```

```python
# Count word frequency in a given text
text = "This is a sample text with several words. This text is just a sample."

# Convert to lowercase and split into words
words = text.lower().split()

# Create a dictionary to count word frequencies
word_frequency = {}
for word in words:
    word = word.strip(".,!?")  # Remove punctuation
    word_frequency[word] = word_frequency.get(word, 0) + 1

# Display the word frequencies
print("Word Frequencies:")
for word, count in word_frequency.items():
    print(f"{word}: {count}")
```

```python
# Inventory system
inventory = {
    "apples": 50,
    "bananas": 30,
    "oranges": 20
}

# Display inventory
print("Current Inventory:")
for item, quantity in inventory.items():
    print(f"{item}: {quantity}")

# Add or update an item
inventory["grapes"] = 40  # Adding grapes
inventory["apples"] += 10  # Restocking apples

# Remove an item
del inventory["oranges"]

# Display updated inventory
print("\nUpdated Inventory:")
for item, quantity in inventory.items():
    print(f"{item}: {quantity}")
```

```python
# Customer orders dataset
orders = [
    {"customer_id": 1, "product_category": "Electronics"},
    {"customer_id": 2, "product_category": "Clothing"},
    {"customer_id": 3, "product_category": "Electronics"},
    {"customer_id": 4, "product_category": "Books"},
    {"customer_id": 5, "product_category": "Clothing"},
    {"customer_id": 6, "product_category": "Electronics"}
]

# Count product category occurrences
category_count = {}
for order in orders:
    category = order["product_category"]
    category_count[category] = category_count.get(category, 0) + 1

# Find the most popular category
most_popular_category = max(category_count, key=category_count.get)

# Display results
print("Product Category Count:")
for category, count in category_count.items():
    print(f"{category}: {count}")

print(f"\nMost Popular Category: {most_popular_category}")
```

```python
# Dictionary to store student grades
grades = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "Diana": 88,
    "Eve": 95
}

# Calculate average grade
total_grades = sum(grades.values())
average_grade = total_grades / len(grades)

# Display grades and average
print("Student Grades:")
for student, grade in grades.items():
    print(f"{student}: {grade}")

print(f"\nAverage Grade: {average_grade:.2f}")
```
